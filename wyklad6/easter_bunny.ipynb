```python
%display typeset

# Macierz wypłat z punktu widzenia św. Mikołaja
#
# Ruchy: św. Mikołaj: R, P, S
#        zajączek: R, P

M = matrix(QQ, [[0, -1], [1, 0], [-1, 1]])
M
```

Problem wyznaczający strategię optymalną dla gracza wierszowego (św. Mikołaja):

$$
	\begin{aligned}
		& x_{0} \to \max \\
		& M^{T} x - \mathbf{1} x_{0} \ge 0, \\
		& \sum_{i=1}^{m} x_{i} = 1, \\
		& x \ge 0, \quad x_{0} \in \mathbb{R}.
	\end{aligned}
$$

```python
A = M.T.augment(vector([-1,-1])).stack(vector([1, 1, 1, 0]))
b = vector([0, 0, 1])
c = vector([0, 0, 0, 1])

A, b, c
```

```python
P = InteractiveLPProblem(A, b, c, x=["x1", "x2", "x3", "x0"], constraint_type=[">=", ">=", "=="], variable_type=[">=", ">=", ">=", ""])
P
```

```python
P.standard_form().run_simplex_method()
```
